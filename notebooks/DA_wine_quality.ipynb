```python
# Exploring outliers identification
# Assuming normal distribution
from scipy import stats

# Z-scores calculation
z_scores = stats.zscore(df[['residual sugar', 'chlorides', 'sulphates']])
outliers_z = (abs(z_scores) > 3).any(axis=1)
outliers_z_df = df[outliers_z]

print("Outliers identified with Z-Score:")
outliers_z_df[['residual sugar', 'chlorides', 'sulphates', 'quality']]
#
```

```python
outliers_z_df['quality'].value_counts()
```

```python
# Calculate Z-scores and remove outliers
def remove_outliers(df, columns):
    # Find the indices of the outliers
    outlier_indices = set()
    for col in columns:
        z_scores = stats.zscore(df[[col]])
        # Identify the outliers for this column
        outliers = np.abs(z_scores) > 3
        # Get the indices of outliers without using flatten
        outlier_indices.update(df.index[np.any(outliers, axis=1)])

    # Remove rows with outliers
    df_cleaned = df.drop(index=outlier_indices)
    return df_cleaned

# Columns to identify outliers
columns = ['residual sugar', 'chlorides', 'sulphates']
df_no_outliers = remove_outliers(df, columns)

# Display the DataFrame without outliers
df_no_outliers.head()
```

## Clustering to explore Data


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Scale the features
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=6, random_state=42)  # Adjust the number of clusters as needed
clusters = kmeans.fit_predict(scaled_X)

# Add the clusters to a copy of the cleaned dataframe
df_clusters = X.copy()
df_clusters['cluster'] = clusters

# Example of visualization using PCA to reduce to 2 dimensions
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_cleaned = pca.fit_transform(scaled_X)

plt.figure(figsize=(10, 7))
plt.scatter(pca_cleaned[:, 0], pca_cleaned[:, 1], c=clusters, cmap='viridis', s=50)
plt.title('Clusters visualized in 2D using PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

## Clustering Internal evaluation


```python
# Comparison of Results with Different Numbers of Clusters
# Elbow method

from sklearn.metrics import silhouette_score

# Range of cluster numbers to try
k_range = range(1, 11)
inertia = []
silhouette_scores = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(scaled_X)
    inertia.append(kmeans.inertia_)
    if k > 1:
        silhouette_scores.append(silhouette_score(scaled_X, clusters))

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')

plt.subplot(1, 2, 2)
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores')

plt.show()
```

```python
# splip df_no_outliers in test and trainning sets
from sklearn.model_selection import train_test_split

X = df_no_outliers.drop('quality', axis=1)
y = df_no_outliers['quality']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

## Lineal Regression

```python
# splip df_no_outliers in test and trainning sets

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, cross_val_predict, KFold
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline


# Create a pipeline with scaling and linear regression
linear_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regression', LinearRegression(fit_intercept=True))
])

# Fit the model on the training data
linear_pipeline.fit(X_train, y_train)

# Evaluate the model on the training set
train_pred = linear_pipeline.predict(X_train)
print("Linear Regression Training Set Evaluation:")
print(f"Mean Squared Error: {mean_squared_error(y_train, train_pred):.2f}")
print(f"R² Score: {r2_score(y_train, train_pred):.2f}")

# Evaluate the model on the test set
test_pred = linear_pipeline.predict(X_test)
print("\nLinear Regression Test Set Evaluation:")
print(f"Mean Squared Error: {mean_squared_error(y_test, test_pred):.2f}")
print(f"R² Score: {r2_score(y_test, test_pred):.2f}")

# Cross-Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_r2_scores = cross_val_score(linear_pipeline, X_train, y_train, cv=kf, scoring='r2')
cv_mse_scores = -cross_val_score(linear_pipeline, X_train, y_train, cv=kf, scoring='neg_mean_squared_error')

print("\nCross-Validated Results:")
print(f"Mean Cross-validated R²: {np.mean(cv_r2_scores):.2f}")
print(f"Mean Cross-validated MSE: {np.mean(cv_mse_scores):.2f}")

# Visualization of Cross-Validation Predictions
cv_predictions = cross_val_predict(linear_pipeline, X_train, y_train, cv=kf)
plt.figure(figsize=(10, 6))
plt.scatter(y_train, cv_predictions, alpha=0.5)
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'r--', lw=2)
plt.xlabel('Actual Quality')
plt.ylabel('Predicted Quality')
plt.title('Linear Regression: Cross-Validated Predictions vs Actual Values')
plt.grid(True)
plt.show()

```

## Polinomial Regresion with Ridge regulization

```python
# splip df_no_outliers in test and trainning sets

from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import Ridge

# Define the pipeline for polynomial regression with ridge regularization
polynomial_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('polynomial', PolynomialFeatures()),
    ('regressor', Ridge(fit_intercept=True))
])

# Define the parameter grid for GridSearch
param_grid = {
    'polynomial__degree': [2, 3, 4, 5],
    'regressor__alpha': np.logspace(-4, 4, 100)
}

# Perform Grid Search with Cross-Validation
grid_search = GridSearchCV(polynomial_pipeline, param_grid, scoring='neg_mean_squared_error', cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)

# Retrieve and evaluate the best model
best_polynomial_model = grid_search.best_estimator_
y_pred = best_polynomial_model.predict(X_test)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("\nPolynomial Regression with Ridge:")
print(f"Best Hyperparameters: {grid_search.best_params_}")
print(f"R² Score: {r2:.2f}")
print(f"Mean Squared Error: {mse:.2f}")

# Cross-Validation on the best model
cv_scores = cross_val_score(best_polynomial_model, X_train, y_train, scoring='neg_mean_squared_error', cv=5, n_jobs=-1)
print(f"Mean Cross-Validation Score: {np.mean(cv_scores):.2f}")
```

```python
#For classification porpuses deleted quality 3 and 8
# splip df_no_outliers in test and trainning sets filter, without quality=3 and quality = 8 dataset
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, label_binarize
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import roc_auc_score, roc_curve, classification_report, confusion_matrix, ConfusionMatrixDisplay
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbalancedPipeline
import numpy as np
import matplotlib.pyplot as plt

# Define the pipeline for logistic regression with SMOTE
logistic_pipeline = ImbalancedPipeline([
    ('smote', SMOTE(sampling_strategy='auto', k_neighbors=2)),
    ('scaler', StandardScaler()),
    ('polynomial', PolynomialFeatures()),
    ('classifier', LogisticRegression(fit_intercept=False, penalty='l2', max_iter=1000))
])

# Define the parameter grid
logistic_param_grid = {
    'polynomial__degree': [2, 3],
    'classifier__C': np.logspace(0, 2, 5)
}

# Perform Grid Search with Stratified K-Folds
logistic_grid_search = GridSearchCV(
    logistic_pipeline,
    logistic_param_grid,
    cv=StratifiedKFold(n_splits=3),
    scoring='neg_log_loss',
    n_jobs=-1
)
logistic_grid_search.fit(X_train, y_train)

# Evaluate the best model
best_logistic_model = logistic_grid_search.best_estimator_
y_pred = best_logistic_model.predict(X_train)

print("\nLogistic Regression:")
print("Best Hyperparameters:", logistic_grid_search.best_params_)

# Ensure target_names matches the unique classes in y_train
classes = sorted(set(y_train))
target_names = [str(cls) for cls in classes]

print("Classification Report:")
print(classification_report(y_train, y_pred, target_names=target_names))

# Compute and plot the Confusion Matrix
conf_matrix = confusion_matrix(y_train, y_pred, labels=classes)
disp = ConfusionMatrixDisplay(conf_matrix, display_labels=target_names)
plt.figure(figsize=(10, 8))
disp.plot(cmap=plt.cm.Blues, values_format='d')
plt.title('Confusion Matrix')
plt.show()

# Plot ROC curve for each class
y_train_bin = label_binarize(y_train, classes=classes)
y_prob = best_logistic_model.predict_proba(X_train)

plt.figure(figsize=(10, 8))
for i in range(len(classes)):
    fpr, tpr, _ = roc_curve(y_train_bin[:, i], y_prob[:, i])
    plt.plot(fpr, tpr, label=f'ROC curve of class {classes[i]} (area = {roc_auc_score(y_train_bin[:, i], y_prob[:, i]):.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves')
plt.legend(loc='best')
plt.show()

```
